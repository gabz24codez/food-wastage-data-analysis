Multivariate analysis:

```python
# Check Total Food Wasted by each Countries in year 2024
plt.figure(figsize=(12,7))
sns.barplot(data=df_2024, x='Country', y='Total Waste (Tons)', ci=None)

plt.title('Total Food Waste by Each Countries in 2024 ')
plt.xlabel('Country')
plt.ylabel('Total Waste (Tons)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Looking at top 10 countries with most food wastages in 2024
topwaste_2024 = df_2024.groupby('Country')['Total Waste (Tons)'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12,7))
sns.barplot(data=topwaste_2024, x='Country', y='Total Waste (Tons)', ci=None)
plt.title('Top 10 Countries Total Food Waste in 2024')
plt.xlabel('Country')
plt.ylabel('Total Waste (Tons)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Looking at top 10 countries with most economic loss in 2024
toploss_2024 = df_2024.groupby('Country')['Economic Loss (Million $)'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12,5))
sns.barplot(data=toploss_2024, x='Country', y='Economic Loss (Million $)', ci=None)
plt.title('Top 10 Countries Economic Loss in 2024 ')
plt.tight_layout()
plt.show()
```

```python
corr_2024 = df_2024.corr
print(corr_2024)
```

```python
# Correlation heatmap 
corr_data = data[['Total Waste (Tons)','Economic Loss (Million $)','Avg Waste per Capita (Kg)','Population (Million)','Household Waste (%)']]
corr_matrix = corr_data.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title("Correlation Matrix on Food Waste")
plt.show()
```

Correlation between Total Waste and Economic Loss has the highest positive value showing that the two category has a strong relation.The positive value (>0) shows that when total food waste increase, economic loss also increase

Avg Waste per Capita, Population and Household Waste hav weak correlation shows that they have low impact 

```python
# Plot to show the positive relation between the economic loss and total food wastage
sns.lmplot(data=df_2024, x='Total Waste (Tons)', y='Economic Loss (Million $)',hue='Food Category', palette='BuPu')
plt.title("Total Waste impact on Economic Loss in 2024")
plt.xlabel("Total Waste (Tons)")
plt.ylabel("Economic Loss (Million $)")
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap graph on Food Categories contribution to Wastage 
foodtypes_table = pd.pivot_table(data=df_2024, index = 'Country', columns = 'Food Category', values = 'Total Waste (Tons)')
plt.figure(figsize=(12, 7))
sns.heatmap(foodtypes_table, annot=True, cmap='BuPu')
plt.title("Heatmap on Wastage by each Food Category in 2024")
plt.xlabel("Food Category")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap graph on Food Categories contribution to Economic Loss
foodtypes_table = pd.pivot_table(data=df_2024, index = 'Country', columns = 'Food Category', values = 'Economic Loss (Million $)')
plt.figure(figsize=(12, 7))
sns.heatmap(foodtypes_table, annot=True, cmap='BuPu')
plt.title("Heatmap on Economic Loss by each Food Category in 2024")
plt.xlabel("Food Category")
plt.ylabel("Country")
plt.show()
```

```python
# Calculate Economic Loss for year 2024 by each Food Category in ascending order
econloss_bycategory = df_2024.groupby("Food Category")["Economic Loss (Million $)"].sum().sort_values(ascending=False)
print(econloss_bycategory)
```

```python
# Bar Chart for total economic loss in year 2024 by food category
plt.figure(figsize=(12,7))
sns.barplot(x=econloss_bycategory.values, y=econloss_bycategory.index, color='skyblue')
plt.title("Total Economic Loss by Food Category in 2024")
plt.xlabel("Economic Loss (Million $)")
plt.ylabel("Food Category")
plt.tight_layout()
plt.show()
```

```python
# Calculate Economic Loss in year 2024 by each Food Category in ascending order
totalwaste_bycategory = df_2024.groupby("Food Category")["Total Waste (Tons)"].sum().sort_values(ascending=False)
print(totalwaste_bycategory)
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x=totalwaste_bycategory.values, y=totalwaste_bycategory.index, color='darkblue')
plt.title("Total Waste by Food Category in 2024")
plt.xlabel("Total Waste (Tons)")
plt.ylabel("Food Category")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart to show the food category percentage contribution in 2024
plt.pie(totalwaste_bycategory.values, labels = totalwaste_bycategory.index, autopct = '%1.1f%%', colors=sns.color_palette('flare'))
plt.title("Total Waste by Food Category in 2024")
plt.tight_layout()
plt.show()
```

Comparing the Total Waste and Economic Loss by Food Category:
The arrangement of ascending food order between the 2 analysis are different shows that the quantity of food waste doesnt equal to economic loss as other factors such as food costs also matters.
